```python
import os
import random
import cv2
import matplotlib.pyplot as plt
```

## Download

```python
def download_unzip(url, output_path):
  os.makedirs(folder, exist_ok=True)
  os.system(f'wget -O {output_path}.zip {download_url}')
  os.system(f'unzip {output_path}.zip -d {output_path}')

download_url = 'https://inf-cv.uni-jena.de/wordpress/wp-content/uploads/2023/03/croatianFishDataset.zip'
folder = '../downloads'
dataset_name = 'croatian_fish'
local_path = f'{folder}/{dataset_name}'
download_unzip(download_url, local_path)
```

## Visualise

```python
dataset_path = f'{local_path}/croatianFishDataset1-0Dir'
total_images = 0
for root, dirs, files in os.walk(dataset_path):
  for file in files:
    if file.endswith('.png') and not file.startswith('.'):
      total_images += 1

print(f'Total images: {total_images}')
```

```python
# Sample Image
species_folder = random.sample([d for d in os.listdir(dataset_path) if os.path.isdir(os.path.join(dataset_path, d))], 1)
image = random.sample([f for f in os.listdir(f'{dataset_path}/{species_folder[0]}') if f.endswith('.png')], 1)
path = os.path.join(dataset_path, species_folder[0], image[0])
print(f'path: {path}')

# # Read annotations
folder_species = {
    "0": "Chromis chromis",
    "1": "Coris julis female",
    "2": "Coris julis male",
    "3": "Diplodus annularis",
    "4": "Diplodus vulgaris",
    "5": "Oblada melanura",
    "6": "Serranus scriba",
    "7": "Spondyliosoma cantharus",
    "8": "Spicara maena",
    "9": "Symphodus melanocercus",
    "10": "Symphodus tinca",
    "11": "Sarpa salpa",
}
species = folder_species[species_folder[0]]

# Show image and associated genus
img = cv2.imread(path)
plt.figure(figsize=(15, 15))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title(species)
plt.show()
```
