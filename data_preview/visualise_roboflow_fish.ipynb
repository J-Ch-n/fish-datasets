```python
from imgann import Sample
import os
```

## Download

```python
def download_unzip(url, output_path):
  os.makedirs(folder, exist_ok=True)
  os.system(f'wget -O {output_path}.zip {url}')
  os.system(f'unzip {output_path}.zip -d {output_path}')

# REPLACE WITH YOUR API KEY
download_url = 'https://public.roboflow.com/ds/eMbxNFju7V?key=<YOUR_API_KEY>'
folder = '../downloads'
dataset_name = 'roboflow_fish'
local_path = f'{folder}/{dataset_name}'
download_unzip(download_url, local_path)
```

## Visualise

```python
format = 'voc'
split = 'train' # choose a split to visualise
dataset_path = os.path.join(local_path, split)
```

```python
Sample.describe_ann(
  data_path=dataset_path,
  ann_path=dataset_path,
  ann_type=format
)

Sample.show_samples(
  data_path=dataset_path,
  ann_path=dataset_path,
  ann_type=format,
  num_of_samples = 1,
  image_shape=[1920, 1080]
)
```
